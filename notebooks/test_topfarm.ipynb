```python
import matplotlib.pyplot as plt

# Przykładowe współrzędne turbin (x, y) w metrach
x = [0, 500, 0, 500]
y = [0, 0, 500, 500]

plt.figure(figsize=(6,6))
plt.scatter(x, y, color='blue', marker='x', s=100, label='Turbiny')
plt.title("Wstępny układ farmy wiatrowej")
plt.xlabel("Pozycja X [m]")
plt.ylabel("Pozycja Y [m]")
plt.grid(True)
plt.legend()
plt.show()
```
