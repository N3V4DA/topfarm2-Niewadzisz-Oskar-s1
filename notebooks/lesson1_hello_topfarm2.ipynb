Zastosowaliśmy zmianę w wykresie przy pomocy AI. Zmiana polegała na dodaniu kolorów dla każdej turbiny, etykiet T1-T4 oraz legendy, co poprawiło czytelność layoutu względem pierwotnej wersji.

```python
import matplotlib.pyplot as plt

x = [0, 500, 1000, 1500]
y = [0, 0, 500, 500]
kolory = ['blue', 'red', 'green', 'yellow']
etykiety = ['T1', 'T2', 'T3', 'T4']

plt.figure(figsize=(8, 6))

for i in range(len(x)):
    plt.scatter(x[i], y[i], color=kolory[i], s=100, label=etykiety[i])
    plt.text(x[i]+20, y[i]+20, etykiety[i])

plt.title("Layout turbin – Zadanie 2.1 z kolorami i etykietami")
plt.xlabel("Pozycja X [m]")
plt.ylabel("Pozycja Y [m]")
plt.grid(True, linestyle='--')
plt.axis('equal')
plt.legend()
plt.show()
```
